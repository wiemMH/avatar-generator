```python
import tensorflow as tf
from keras.models import load_model
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = tf.keras.models.load_model("generator.h5", custom_objects=None, compile=True, options=None)
```

```python
model.compile()
```

```python
batch_size = 5
latent_dim = 128
random_latent_vectors = tf.random.normal(shape=(batch_size, latent_dim))
generated_images = model.predict(random_latent_vectors)
```

```python
fig = plt.figure(figsize=[16,8])
for i in range(batch_size):
    ax = plt.subplot(4,8,i+1)
    plt.imshow((generated_images*255).astype(np.uint8)[i])
    plt.axis("off")
```

```python
plt.axis("off")
plt.imshow((generated_images*255).astype("int32")[i])
```

```python
def image_generator(batch_size, latent_dim):
    random_latent_vectors = tf.random.normal(shape=(batch_size,latent_dim))
    generated_images = model.predict(random_latent_vectors)
    return generated_images
```
